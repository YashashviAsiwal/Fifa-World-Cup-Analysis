**FIFA WORLD CUP**

The FIFA World Cup is the premier international football tournament, first held in 1930. Organized by FIFA, it occurs every four years and features 32 teams competing in a month-long event. The tournament has seen iconic moments and legendary players, with Brazil holding the most titles (five). The World Cup is a global spectacle that influences culture, economy, and football development, drawing immense viewership and uniting fans across the world. Starting in 2026, the tournament will expand to include 48 teams, further broadening its global impact.

COUNTRIES THAT WON THE CUP

```python
# Counting podium finishes
gold = world_cup["Winner"]
silver = world_cup["Runners-Up"]
bronze = world_cup["Third"]

# Creating data frames from value counts and renaming columns
gold_count = pd.DataFrame(gold.value_counts()).reset_index()
gold_count.columns = ['Country', 'WINNER']

silver_count = pd.DataFrame(silver.value_counts()).reset_index()
silver_count.columns = ['Country', 'SECOND']

bronze_count = pd.DataFrame(bronze.value_counts()).reset_index()
bronze_count.columns = ['Country', 'THIRD']

# Joining counts into a single DataFrame
podium_count = gold_count.set_index('Country').join(silver_count.set_index('Country'), how='outer').join(bronze_count.set_index('Country'), how='outer')
podium_count = podium_count.fillna(0).astype('int64')

# Sorting values for better visualization
podium_count = podium_count.sort_values(by=['WINNER', 'SECOND', 'THIRD'], ascending=False)

# Plotting
podium_count.plot(
    y=['WINNER', 'SECOND', 'THIRD'], kind="bar", 
    color=['gold', 'silver', 'brown'], figsize=(15, 6), fontsize=14,
    width=0.8, align='center'
)
plt.xlabel('Countries')
plt.ylabel('Number of Podium Finishes')
plt.title('Number of Podium Finishes by Country')
plt.grid(axis='y', linestyle='--', linewidth=0.7)  # Add a horizontal grid for better readability
plt.legend(title='Podium Position')

# Display the plot
plt.show()
```

NUMBER OF GOALS PER COUNTRY

```python
# Extract relevant columns and drop NA values
home = matches[['Home Team Name', 'Home Team Goals']].dropna()
away = matches[['Away Team Name', 'Away Team Goals']].dropna()

# Rename columns to have a common schema
home = home.rename(columns={'Home Team Name': 'countries', 'Home Team Goals': 'goals'})
away = away.rename(columns={'Away Team Name': 'countries', 'Away Team Goals': 'goals'})

# Combine home and away goals into one DataFrame using pd.concat
goal_per_country = pd.concat([home, away])

# Convert goals to integer type
goal_per_country['goals'] = goal_per_country['goals'].astype('int64')

# Group by country and sum goals
goal_per_country = goal_per_country.groupby('countries')['goals'].sum().sort_values(ascending=False)

# Plot the top 10 countries
goal_per_country.head(10).plot(kind='bar', figsize=(12, 6), fontsize=14, color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Number of Goals')
plt.title('Top 10 Countries by Number of Goals')
plt.grid(axis='y', linestyle='--', linewidth=0.7)  # Add horizontal grid for readability

# Display the plot
plt.show()
```

MATCH OUTCOME BY HOME AND AWAY TEAMS

```python
def get_labels(matches):
    if matches['Home Team Goals'] > matches['Away Team Goals']:
        return 'Home Team Win'
    if matches['Home Team Goals'] < matches['Away Team Goals']:
        return 'Away Team Win'
    return 'DRAW'
```

```python
matches['outcome'] = matches.apply(lambda x: get_labels(x), axis=1)
```

```python
matches.head()
```

```python
mt = matches['outcome'].value_counts()
mt
```

```python
plt.figure(figsize = (6,6))

mt.plot.pie(autopct = "%1.0f%%", colors = sns.color_palette('winter_r'), shadow = True)

c = plt.Circle((0,0), 0.4, color =  'white')
plt.gca().add_artist(c)
plt.title('Match Outcomes by Home and Away Teams')
plt.show()
```
